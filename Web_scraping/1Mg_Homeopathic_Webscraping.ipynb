```python
# Importing libraries
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
```

### Table_1

```python
# Table_1: Medicine_Name

Name=[]
Sizes=[]
MRPs=[]
Prices=[]
URLs=[]

page=1

while True:
    header={
        'Origin': 'https://www.1mg.com',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'
    }
    url="https://www.1mg.com/categories/homeopathy/homeopathy-medicines-106?filter=true&pageNumber="+str(page)
    res = requests.get(url=url,headers=header)
    # print(res.status_code)

    if res.status_code==404:
        print("break")
        break
    else:
        # print('else')
        obj=soup(res.content,'html.parser')
        Box=obj.find_all('div',{'class':'style__product-box___liepi'})
        for i in Box:
            name=i.find('div', {'class':'style__pro-title___2QwJy'})
            if name:
                Name.append(name.text)
            else:
                Name.append("Na")

            size=i.find('div',{'class':'style__pack-size___2JQG7'})
            if size:
                Sizes.append(size.text)
            else:
                Sizes.append("Na")

            mrps=i.find('span',{'class':'style__discount-price___25Bya'})
            if mrps:
                MRPs.append(mrps.text)
            else:
                MRPs.append("NA")

            price=i.find('div',{'class':'style__price-tag___cOxYc'})
            if price:
                Prices.append(price.text)
            else:
                Prices.append("NA")
            
            mg_url=i.find('a',{'class':'style__product-link___UB_67'})
            if mg_url:
                b="https://www.1mg.com/"+mg_url.get('href')
                URLs.append(b)
            else:
                URLs.append("NA")
        print(page,"page=>",end=" ")
    page+=1
#     print(len(URLs),end=" ")
#     print(len(Prices),end=" ")
#     print(len(Sizes),end=" ")
#     print(len(MRPs),end=" ")
#     print(len(Name))


data={"name":Name,"size_of_bottle":Sizes,"MRPs":MRPs,"selling_price":Prices,"1mg_url":URLs}
# print(data)
dF1=pd.DataFrame(data)
df1.head()
```

```python
# getting the link of all the products

box_link=[]
for i in range(0,100):
    var={"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"}
    url ="https://www.1mg.com/categories/homeopathy/homeopathy-medicines-106?filter=true&pageNumber="+str(i)
    response=requests.get(url, headers=var)
    soup=BeautifulSoup(response.text, "lxml")
 
    for i in soup.find_all("a"):
        com_link="https://www.1mg.com"+ str(i.get("href"))
        box_link.append(com_link)
#print(box_link)
print(len(box_link))    
```

```python
# to check and slice the useful link

print(box_link.index("https://www.1mg.com/otc/sbl-asarum-europaeum-dilution-30-ch-otc345941"))
print(box_link[0:100])
```

```python
# Table_2: Medicine_Details

URLs=[]
Names=[]
Brands=[]
Desc=[]
Benefits=[]
Ingredients=[]
Ratings=[]
Reviews=[]

for url in box_link[0:100]:
    var={'Origin': 'https://www.1mg.com',
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"}
    
    response = requests.get(url, headers=var)
    if response.status_code==404:
        break
    else:
        URLs.append(url)
        
        soup = BeautifulSoup(response.text, "lxml")

        name = soup.find("h1", {"class": "ProductTitle__product-title___3QMYH"})
        if name:
            Names.append(name.text)
        else:
            Names.append("NA")

        brand=soup.find("div", {"class":"ProductTitle__manufacturer___sTfon"})
        if brand:
            Brands.append(brand.text)
        else:
            Brands.append("NA")

    #     desc=soup.find_all("div", {"class":"ProductDescription__product-description___1PfGf"})
    #     if desc:
    #         Desc.append(desc.text)
    #     else:
    #         Desc.append("NA")

        benefit=soup.find("div", {"class":"ProductHighlights__highlights-text___dc-WQ"})
        if benefit:
            Benefits.append(benefit.text)
        else:
            Benefits.append("NA")

    #     ingredient=soup.find_all("", {"class":""})
    #     if ingredient:
    #         Ingredients.append(ingredient.text)
    #     else:
    #         Ingredients.append("NA")

        rating=soup.find("div", {"class":"RatingDisplay__ratings-container___3oUuo"})
        if rating:
            Ratings.append(rating.text)
        else:
            Ratings.append("NA")

        review=soup.find("span", {"class":"RatingDisplay__ratings-header___ZNj5b"})
        if review:
            Reviews.append(review.text)
        else:
            Reviews.append("NA")

#     print(len(URLs),end=" ")
#     print(len(Names),end=" ")
#     print(len(Brands),end=" ")
#     print(len(Benefits),end=" ")
#     print(len(Ratings),end=" ")
#     print(len(Reviews),end=" ")
    
dic={
     "urls":URLs,
     "name": Names, 
     "brand_name":Brands,
     "key_benefits":Benefits,  
     "rating":Ratings, 
     "number_of_rating": Reviews
    }
df2=pd.DataFrame(dic)
df2.head()
```
